# Fitts Law

Theoretical description of the modelling framework: Fitts Law Task.

Fitts law is a gauge for one's confortability with their computer/testing environment. Fitts law has been rigorously tested & as such a candidates deviation from the expected response time (RT).


It may be worth modelling fitts law with a series of flexible basis functions or ridge regression, to gauge how the size of the components or position on the screen plays a role in the candidates ability to complete the task, however a simple average deviation from expected response time (in miliseconds) should be sufficient as a proxy for thier comfortability with thier computer.

A parameter can thus be added to each model (NBack & WCST):

$$: 
\begin{equation}
    f_p = \frac{1}{n} \sum_{j=1}^{n} (RT - \mathbb{E}[RT])
  \end{equation}
$$


For each $participant \ p$. Not the sign $+/-$ of $f_p$ is informative.

# Corsi Block Span

Corsi Block span can offer a gauge for working memory. An individuals longest block span can be used as a proxy for either _attentiveness_ or _working memory_.


$$: 
\begin{equation}
    c_p := \text{participant p's longest corsi plan}
  \end{equation}
$$

# Navon

A measurement of how individuals process global and local information. The experiments offers the following data:

1. The letter. For example, Hs stands for a big H made of small letters S
2. The level where the target is, or none if there is no target letter at all
3. Same as column 2, but in numbers. 0: no target. 1: target at local level. 2: target at global level
4. Status. 1: correct, 2: wrong response, 3: too slow
5. Reaction time in milliseconds


This infomation could be summarised in a number of ways, a simple configuration is simply to use the proportion of correct answers of both target types: global and local information.


$$
\begin{eqnarray} 
    \delta_p^g &:=& \text{proportion of global processing examples identified correctly} \\
    \delta_p^l &:=& \text{proportion of local processing examples identified correctly} 
\end{eqnarray}
$$


For partipant $p$.
